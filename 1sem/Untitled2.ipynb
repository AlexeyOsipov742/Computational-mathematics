```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

**1) Классическая полиномиальная интерполяция по Ньютону**

```python
years = np.array([1950, 1960, 1970, 1980, 1990, 2000])
population = np.array([546815000, 667070000, 818315000, 981235000, 1135185000, 1262645000])
```

```python
def divided_differences(years, population):
    """Рекурсивно вычисляет таблицу разделённых разностей"""
    n = len(population)
    div_diff = np.zeros((n, n))
    div_diff[:, 0] = population  # Первая колонка — значения функции

    for j in range(1, n):
        for i in range(n - j):
            div_diff[i, j] = (div_diff[i + 1, j - 1] - div_diff[i, j - 1]) / (years[i + j] - years[i])

    return div_diff[0]  # Возвращаем первую строку с коэффициентами

def newton_polynomial(x, x_points, coef):
    """Вычисление значения полинома Ньютона в точке x"""
    n = len(coef)
    value = coef[0]
    for i in range(1, n):
        term = coef[i]
        for j in range(i):
            term *= (x - x_points[j])
        value += term
    return value

# Вычисляем разделённые разности
coefficients = divided_differences(years, population)

# Полином для предсказания (например, для 1965 года)
year_to_predict = 2010
predicted_population = newton_polynomial(year_to_predict, years, coefficients)
print(predicted_population)
```

```python

x_vals = np.linspace(1940, 2010, 500)  # Годы от 1940 до 2010
y_vals = [newton_polynomial(x, years, coefficients) for x in x_vals]

# Построение графика
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label="Интерполяционный полином", color="blue")
plt.scatter(years, population, color="red", label="Исходные точки", zorder=5)
plt.title("Интерполяция населения Китая (1950-2000)")
plt.xlabel("Год")
plt.ylabel("Население (чел)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
