**Цель**:  построить графики абсолютной погрешности каждого метода в зависимости от шага численного дифференцирования $h_n = \frac{2}{2^n}$ для $n=1..21$ для набора из 5 функций:
$$sin(x^2), \space\space cos(sin(x)), \space\space exp(sin(cos(x)), \space\space ln(x+3),  \space\space \sqrt{x+3}.$$ <br>

Возьмём произвольную точку $x$ и сравним производную в этой точке, полученную аналитически, и с помощью формул 1-5: 
1. $$\frac{f(x+h) - f(x)}{h}$$
2. $$\frac{f(x) - f(x-h)}{h}$$
3. $$\frac{f(x+h) - f(x-h)}{h}$$
4. $$\frac{4}{3}\frac{f(x+h)-f(x-h)}{2h} - \frac{1}{3}\frac{f(x+2h)-f(x-2h)}{4h}$$
5. $$\frac{3}{2}\frac{f(x+h)-f(x-h)}{2h} -\frac{3}{5}\frac{f(x+2h)-f(x-2h)}{4h} + \frac{1}{10}\frac{f(x+3h)-f(x-3h)}{6h}$$
 
Результатом исследования будут графики абсолютной погрешности - разности аналитического и численного расчета производных в зависимости от размера $h$ в численном методе.

```python
#Используемые библиотеки
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

#Функция численной производной
def derivative_of_f(x, h, f, n):
    if n == 1:
        return ( (f(x+h)-f(x))/h )
    elif n == 2:
        return ( (f(x)-f(x-h))/h )
    elif n == 3:
        return ( (f(x+h)-f(x-h))/(2*h) )
    elif n == 4:
        return ( (4/3*(f(x+h)-f(x-h))/(2*h)) - (1/3*(f(x+2*h)-f(x-2*h))/(4*h)) )
    elif n==5:
        return ( (3/2*(f(x+h)-f(x-h))/(2*h)) - (3/5*(f(x+2*h)-f(x-2*h))/(4*h)) + (1/10*(f(x+3*h)-f(x-3*h))/(6*h)) )
```

```python
#Выбранная точка
x = np.pi/6  #np.pi/2
n = np.arange(1, 22, 1)
h = 2/2**n
```

### 1) $sin(x^2)$

```python
#Сама функция 
def f(x):
    return np.sin(x**2)
#Её аналитическая производная
def true_derivative_of_f(x):
    return 2*x*np.cos(x**2)
```

```python
plt.figure(figsize=[13, 6], dpi=350)
plt.plot(h, np.array(abs(true_derivative_of_f(x) - derivative_of_f(x, h, f, 1))), '.-', label='y1')
plt.plot(h, np.array(abs(true_derivative_of_f(x) - derivative_of_f(x, h, f, 2))), '.-', label='y2')
plt.plot(h, np.array(abs(true_derivative_of_f(x) - derivative_of_f(x, h, f, 3))), '.-', label='y3')
plt.plot(h, np.array(abs(true_derivative_of_f(x) - derivative_of_f(x, h, f, 4))), '.-', label='y4')
plt.plot(h, np.array(abs(true_derivative_of_f(x) - derivative_of_f(x, h, f, 5))), '.-', label='y5')
    
plt.xlabel("$h$")
plt.ylabel(" true_der - der(h)")
plt.xscale('log')
plt.yscale('log')
plt.legend()
plt.title("Абсолютная погрешность методов для f1 = sin(x^2)")
plt.grid()
plt.show()
```

### 2) $cos(sin(x))$

```python
def f(x):
    return np.cos(np.sin(x))
    
def true_derivative_of_f(x):
    return -np.cos(x)*np.sin(np.sin(x))
```

```python
plt.figure(figsize=[13, 6], dpi=350)
plt.plot(h, np.array(abs(true_derivative_of_f(x) - derivative_of_f(x, h, f, 1))), '.-', label='y1')
plt.plot(h, np.array(abs(true_derivative_of_f(x) - derivative_of_f(x, h, f, 2))), '.-', label='y2')
plt.plot(h, np.array(abs(true_derivative_of_f(x) - derivative_of_f(x, h, f, 3))), '.-', label='y3')
plt.plot(h, np.array(abs(true_derivative_of_f(x) - derivative_of_f(x, h, f, 4))), '.-', label='y4')
plt.plot(h, np.array(abs(true_derivative_of_f(x) - derivative_of_f(x, h, f, 5))), '.-', label='y5')
    
plt.xlabel("$h$")
plt.ylabel(" true_der - der(h)")
plt.xscale('log')
plt.yscale('log')
plt.legend()
plt.title("Абсолютная погрешность методов для f2 = cos(sin(x))")
plt.grid()
plt.show()
```

### 3) $exp(sin(cos(x)))$

```python
def f(x):
    return np.exp(np.sin(np.cos(x)))
    
def true_derivative_of_f(x):
    return -np.sin(x) * np.cos(np.cos(x)) * np.exp(np.sin(np.cos(x)))
```

```python
plt.figure(figsize=[13, 6], dpi=350)
plt.plot(h, np.array(abs(true_derivative_of_f(x) - derivative_of_f(x, h, f, 1))), '.-', label='y1')
plt.plot(h, np.array(abs(true_derivative_of_f(x) - derivative_of_f(x, h, f, 2))), '.-', label='y2')
plt.plot(h, np.array(abs(true_derivative_of_f(x) - derivative_of_f(x, h, f, 3))), '.-', label='y3')
plt.plot(h, np.array(abs(true_derivative_of_f(x) - derivative_of_f(x, h, f, 4))), '.-', label='y4')
plt.plot(h, np.array(abs(true_derivative_of_f(x) - derivative_of_f(x, h, f, 5))), '.-', label='y5')
    
plt.xlabel("$h$")
plt.ylabel(" true_der - der(h)")
plt.xscale('log')
plt.yscale('log')
plt.legend()
plt.title("Абсолютная погрешность методов для  f3 = exp(sin(cos(x)))")
plt.grid()
plt.show()
```

### 4) $ln(x+3)$

```python
def f(x):
    return  np.log(x+3)
    
def true_derivative_of_f(x):
    return 1/(x+3)
```

```python
plt.figure(figsize=[13, 6], dpi=350)
plt.plot(h, np.array(abs(true_derivative_of_f(x) - derivative_of_f(x, h, f, 1))), '.-', label='y1')
plt.plot(h, np.array(abs(true_derivative_of_f(x) - derivative_of_f(x, h, f, 2))), '.-', label='y2')
plt.plot(h, np.array(abs(true_derivative_of_f(x) - derivative_of_f(x, h, f, 3))), '.-', label='y3')
plt.plot(h, np.array(abs(true_derivative_of_f(x) - derivative_of_f(x, h, f, 4))), '.-', label='y4')
plt.plot(h, np.array(abs(true_derivative_of_f(x) - derivative_of_f(x, h, f, 5))), '.-', label='y5')
    
plt.xlabel("$h$")
plt.ylabel(" true_der - der(h)")
plt.xscale('log')
plt.yscale('log')
plt.legend()
plt.title("Абсолютная погрешность методов для f4 = ln(x+3)")
plt.grid()
plt.show()
```

### 5) $\sqrt{x+3}$

```python
def f(x):
    return  np.sqrt(x+3)
    
def true_derivative_of_f(x):
    return 1/(2*np.sqrt(x+3))
```

```python
plt.figure(figsize=[13, 6], dpi=350)
plt.plot(h, np.array(abs(true_derivative_of_f(x) - derivative_of_f(x, h, f, 1))), '.-', label='y1')
plt.plot(h, np.array(abs(true_derivative_of_f(x) - derivative_of_f(x, h, f, 2))), '.-', label='y2')
plt.plot(h, np.array(abs(true_derivative_of_f(x) - derivative_of_f(x, h, f, 3))), '.-', label='y3')
plt.plot(h, np.array(abs(true_derivative_of_f(x) - derivative_of_f(x, h, f, 4))), '.-', label='y4')
plt.plot(h, np.array(abs(true_derivative_of_f(x) - derivative_of_f(x, h, f, 5))), '.-', label='y5')
    
plt.xlabel("$h$")
plt.ylabel(" true_der - der(h)")
plt.xscale('log')
plt.yscale('log')
plt.legend()
plt.title("Абсолютная погрешность методов для  f5 = sqrt(x+3)")
plt.grid()
plt.show()
```

Графики показывают, какая из формул лучше всего вычисляет производную. Также по ним можно заметить, что почти у всех методов есть некий оптимальный шаг, что связано с машинной погрешностью. Таким образом, уменьшать $h$ до бесконечности нет смысла.
